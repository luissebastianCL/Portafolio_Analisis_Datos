# Lección 1: Método científico y estadística

1. **Definición del problema**:

Queremos analizar los hábitos saludables en jóvenes universitarios, centrándonos en tres variables principales:
- Sueño (horas de descanso por día).
- Alimentación (frecuencia de consumo de frutas/verduras y comida rápida).
- Actividad física (minutos de ejercicio semanal).

Problema de investigación:

 **¿Existen patrones estadísticamente significativos en los hábitos de sueño, alimentación y actividad física de los jóvenes universitarios que puedan influir en su bienestar general?**


2. **Hipótesis**:

- **Hipótesis nula (H₀):** No existen diferencias significativas en los hábitos saludables de los estudiantes universitarios; los patrones observados son producto del azar.

- **Hipótesis alternativa (H₁):** Sí existen diferencias significativas en los hábitos saludables, y estas están asociadas a factores como género, edad o nivel socioeconómico.




3. **Variables relevantes**:

**Cuantitativas**:
- Horas de sueño (numérica continua).
- Minutos de ejercicio semanal (numérica continua).

**Categóricas**:
- Género (masculino, femenino, otro).
- Nivel socioeconómico (bajo, medio, alto).
- Frecuencia de consumo de frutas/verduras (alta, media, baja).
- Frecuencia de consumo de comida rápida (alta, media, baja).


4. **Metodología**:

- **Enfoque:** Método científico aplicado con diseño observacional y análisis estadístico inferencial.

- **Dataset:** Se simulará un conjunto de datos con al menos 100 registros.

- **Herramientas:** Python (Pandas, NumPy, Matplotlib, Scipy).

- **Diseño preliminar:** Dataset con variables cuantitativas y cualitativas, análisis descriptivo inicial y preparación para inferencia estadística en las siguientes lecciones.


5. **Código inicial** (Preparación Data Set):

```python
# Simulación inicial de dataset con variables relevantes

import pandas as pd
import numpy as np

# Definimos número de registros
n = 120

# Simulación de variables
np.random.seed(42)  # reproducibilidad

# Horas de sueño (media 7, desviación 1.5)
horas_sueno = np.round(np.random.normal(loc=7, scale=1.5, size=n), 2)


# Minutos de ejercicio semanal (media 150, desviación 60)
ejercicio = np.round(np.random.normal(loc=150, scale=60, size=n), 2)

# Género (categoría)
genero = np.random.choice(['Masculino', 'Femenino', 'Otro'], size=n, p=[0.45, 0.45, 0.10])

# Nivel socioeconómico
nivel_socio = np.random.choice(['Bajo', 'Medio', 'Alto'], size=n, p=[0.3, 0.5, 0.2])

# Frecuencia de consumo de frutas/verduras
frutas = np.random.choice(['Alta', 'Media', 'Baja'], size=n, p=[0.4, 0.4, 0.2])

# Frecuencia de consumo de comida rápida
comida_rapida = np.random.choice(['Alta', 'Media', 'Baja'], size=n, p=[0.3, 0.5, 0.2])

# Construcción del DataFrame
df = pd.DataFrame({
    'Horas_sueno': horas_sueno,
    'Ejercicio_min_sem': ejercicio,
    'Genero': genero,
    'Nivel_socio': nivel_socio,
    'FrutasVerduras': frutas,
    'ComidaRapida': comida_rapida
})

# Vista preliminar
print(df.head())


```

# Lección 2: Probabilidad y estadística

Modelar eventos estadísticos, definir el espacio muestral y diseñar un plan de muestreo con al menos 100 registros.

**Detalle del análisis a desarrollar**:

1. **Eventos aleatorios y árboles de probabilidad:**

Ejemplo: seleccionar un estudiante al azar y observar dos variables: género y nivel socioeconómico.

2. **Tipo de muestreo:**

Usaremos muestreo aleatorio simple para garantizar representatividad.
Más adelante podemos comparar con muestreo estratificado (por género o nivel socioeconómico).

3. **Simulación del diseño muestral:**

Ya tenemos un dataset con 120 registros. Para esta lección, seleccionaremos una muestra aleatoria de 100 estudiantes.

4. **Cálculo de probabilidades básicas:**

- Probabilidad de que un estudiante duerma más de 8 horas.
- Probabilidad de que un estudiante sea de nivel socioeconómico alto.
- Probabilidad conjunta de que un estudiante sea mujer y tenga consumo alto de frutas.
- Probabilidad complementaria de que un estudiante no consuma comida rápida con frecuencia alta.


```python
import pandas as pd
import numpy as np

# Usamos el dataset creado en la Lección 1 (df)
# Seleccionamos una muestra aleatoria de 100 estudiantes
muestra = df.sample(n=100, random_state=42)

# Probabilidad de que un estudiante duerma más de 8 horas
p_sueno_mas8 = (muestra['Horas_sueno'] > 8).mean()

# Probabilidad de nivel socioeconómico alto
p_nivel_alto = (muestra['Nivel_socio'] == 'Alto').mean()

# Probabilidad conjunta: mujer y consumo alto de frutas

p_mujer_frutas = ((muestra['Genero'] == 'Femenino') &
                  (muestra['FrutasVerduras'] == 'Alta')).mean()

# Probabilidad complementaria: NO consumo alto de comida rápida
p_no_comida_rapida = (muestra['ComidaRapida'] != 'Alta').mean()

# Resultados con dos decimales
print("Probabilidad de dormir > 8 horas:", round(p_sueno_mas8, 2))
print("Probabilidad de nivel socioeconómico alto:", round(p_nivel_alto, 2))
print("Probabilidad conjunta mujer y frutas altas:", round(p_mujer_frutas, 2))
print("Probabilidad complementaria NO comida rápida alta:", round(p_no_comida_rapida, 2))

```

**Interpretación preliminar**:

- Estas probabilidades nos permiten entender la distribución de hábitos en la muestra.
- El muestreo aleatorio simple asegura que cada estudiante tiene la misma probabilidad de ser seleccionado.
- En próximas lecciones, estas probabilidades se conectarán con distribuciones de probabilidad y pruebas de hipótesis.


# Lección 3: Distribución de probabilidad

Aplicar distribuciones adecuadas según el tipo de variable y contexto, justificando la elección y calculando probabilidades.

 **Identificación de variables y distribuciones**:

- Horas de sueño → variable continua → se ajusta a una distribución normal (aprox. media 7, desviación 1.5).

- Ejercicio semanal (minutos) → variable continua → también se ajusta a una distribución normal (aprox. media 150, desviación 60).

- Consumo de frutas/verduras → variable categórica → se puede modelar con una distribución multinomial.

- Consumo de comida rápida → variable categórica → igualmente multinomial.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Histograma y curva normal para horas de sueño
sns.histplot(df['Horas_sueno'], bins=15, kde=True, color='skyblue')
mu_sueno, sigma_sueno = df['Horas_sueno'].mean(), df['Horas_sueno'].std()
x = np.linspace(df['Horas_sueno'].min(), df['Horas_sueno'].max(), 100)
plt.plot(x, norm.pdf(x, mu_sueno, sigma_sueno)*len(df['Horas_sueno'])*(df['Horas_sueno'].max()-df['Horas_sueno'].min())/15, 'r')
plt.title(f"Distribución de Horas de Sueño (μ={mu_sueno:.2f}, σ={sigma_sueno:.2f})")
plt.show()

# Histograma y curva normal para ejercicio semanal
sns.histplot(df['Ejercicio_min_sem'], bins=15, kde=True, color='lightgreen')
mu_ej, sigma_ej = df['Ejercicio_min_sem'].mean(), df['Ejercicio_min_sem'].std()
x = np.linspace(df['Ejercicio_min_sem'].min(), df['Ejercicio_min_sem'].max(), 100)
plt.plot(x, norm.pdf(x, mu_ej, sigma_ej)*len(df['Ejercicio_min_sem'])*(df['Ejercicio_min_sem'].max()-df['Ejercicio_min_sem'].min())/15, 'r')
plt.title(f"Distribución de Ejercicio Semanal (μ={mu_ej:.2f}, σ={sigma_ej:.2f})")
plt.show()

# Probabilidades con distribución normal
# Ejemplo: probabilidad de dormir más de 8 horas
p_sueno_mas8 = 1 - norm.cdf(8, mu_sueno, sigma_sueno)

# Ejemplo: probabilidad de hacer más de 180 minutos de ejercicio semanal
p_ej_mas180 = 1 - norm.cdf(180, mu_ej, sigma_ej)

print("Probabilidad de dormir > 8 horas:", round(p_sueno_mas8, 2))
print("Probabilidad de ejercicio > 180 min:", round(p_ej_mas180, 2))
```

**Interpretación preliminar**:

- Los histogramas muestran que las variables cuantitativas se ajustan razonablemente a una distribución normal.
- Las probabilidades calculadas permiten responder preguntas prácticas:
- ¿Qué proporción de estudiantes duerme más de 8 horas?
- ¿Qué proporción realiza más de 180 minutos de ejercicio semanal?
- Las variables categóricas (alimentación) se modelarán más adelante con tablas de frecuencia y proporciones.


# Lección 4: Distribución Muestral y TLC

Aplicar el concepto de distribución muestral de la media y verificar empíricamente el Teorema del Límite Central (TLC).

La idea es lo siguiente:

- Generar distribuciones muestrales de la media a partir del dataset simulado.
- Verificar el TLC: al aumentar el tamaño muestral, la distribución de las medias tiende a una normal.
- Comparar la distribución poblacional con la muestral.
- Analizar la dispersión de la media según tamaño muestral.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Función para generar distribución muestral de la media
def distribucion_muestral(variable, n_muestra, repeticiones=1000):
    medias = []
    for _ in range(repeticiones):
        muestra = np.random.choice(variable, size=n_muestra, replace=True)
        medias.append(np.mean(muestra))
    return np.round(medias, 2)

# Ejemplo con horas de sueño
medias_sueno_30 = distribucion_muestral(df['Horas_sueno'], n_muestra=30)
medias_sueno_50 = distribucion_muestral(df['Horas_sueno'], n_muestra=50)
medias_sueno_80 = distribucion_muestral(df['Horas_sueno'], n_muestra=80)

# Graficamos las distribuciones muestrales
plt.figure(figsize=(12,6))

sns.histplot(medias_sueno_30, bins=20, kde=True, color='skyblue', label='n=30')
sns.histplot(medias_sueno_50, bins=20, kde=True, color='lightgreen', label='n=50')
sns.histplot(medias_sueno_80, bins=20, kde=True, color='salmon', label='n=80')

plt.title("Distribución muestral de la media (Horas de sueño)")
plt.xlabel("Media muestral")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# Comparación con distribución poblacional
plt.figure(figsize=(10,5))
sns.histplot(df['Horas_sueno'], bins=20, kde=True, color='gray', label='Población')
sns.histplot(medias_sueno_50, bins=20, kde=True, color='blue', label='Muestral n=50')
plt.title("Comparación distribución poblacional vs muestral")
plt.legend()
plt.show()

# Dispersión de la media según tamaño muestral
print("Desviación estándar de medias (n=30):", round(np.std(medias_sueno_30), 2))
print("Desviación estándar de medias (n=50):", round(np.std(medias_sueno_50), 2))
print("Desviación estándar de medias (n=80):", round(np.std(medias_sueno_80), 2))
```

**Interpretación preliminar**:

- A medida que el tamaño muestral aumenta, la dispersión de las medias disminuye, confirmando el TLC.
- La distribución de las medias tiende a una forma normal, incluso si la población original no es perfectamente normal.
- Esto valida que podemos usar la distribución normal como aproximación para inferencia estadística en las siguientes lecciones.


# Lección 5: Inferencia e intervalos de confianza

Estimar parámetros poblacionales mediante intervalos de confianza y analizar cómo el tamaño muestral afecta el ancho del intervalo.

Lo que se busca es:

- Calcular intervalos de confianza para la media de dos variables cuantitativas:
- Horas de sueño.
- Minutos de ejercicio semanal.
- Usar distintos niveles de confianza (90%, 95%, 99%).
- Evaluar el impacto del tamaño muestral sobre el ancho del intervalo.
- Redactar interpretaciones precisas.


```python
import numpy as np
import scipy.stats as st

# Función para calcular intervalos de confianza
def intervalo_confianza(data, confianza=0.95):
    n = len(data)
    media = np.mean(data)
    sem = st.sem(data)  # error estándar de la media
    margen = sem * st.t.ppf((1 + confianza) / 2, n - 1)
    return round(media, 2), round(media - margen, 2), round(media + margen, 2)

# Ejemplo con Horas de sueño
for nivel in [0.90, 0.95, 0.99]:
    media, li, ls = intervalo_confianza(df['Horas_sueno'], confianza=nivel)
    print(f"Intervalo {int(nivel*100)}% Horas de sueño: Media={media}, LI={li}, LS={ls}")

# Ejemplo con Ejercicio semanal
for nivel in [0.90, 0.95, 0.99]:
    media, li, ls = intervalo_confianza(df['Ejercicio_min_sem'], confianza=nivel)
    print(f"Intervalo {int(nivel*100)}% Ejercicio semanal: Media={media}, LI={li}, LS={ls}")

# Impacto del tamaño muestral

muestra_peq = df['Horas_sueno'].sample(30, random_state=42)
muestra_grande = df['Horas_sueno'].sample(80, random_state=42)

media_peq, li_peq, ls_peq = intervalo_confianza(muestra_peq, confianza=0.95)
media_grande, li_grande, ls_grande = intervalo_confianza(muestra_grande, confianza=0.95)

print("\nImpacto del tamaño muestral (95% confianza):")
print(f"Muestra pequeña (n=30): Media={media_peq}, LI={li_peq}, LS={ls_peq}")
print(f"Muestra grande (n=80): Media={media_grande}, LI={li_grande}, LS={ls_grande}")

```

**Interpretación preliminar**:

- Los intervalos de confianza muestran un rango dentro del cual se espera que esté la media poblacional.
- A mayor nivel de confianza (99%), el intervalo se hace más amplio, reflejando mayor seguridad pero menos precisión.
- A mayor tamaño muestral, el intervalo se hace más estrecho, lo que significa estimaciones más precisas.
- Esto confirma la importancia de un buen diseño muestral para obtener conclusiones confiables


# Lección 6: Test de significancia

Realizar pruebas de hipótesis para validar afirmaciones sobre la población, calcular el valor-p y explicar errores tipo I y II.

Prueba de hipótesis para la media de horas de sueño:
- H₀: La media de horas de sueño = 7 horas.
- H₁: La media de horas de sueño ≠ 7 horas.

Prueba de hipótesis para la media de ejercicio semanal:
- H₀: La media de ejercicio semanal = 150 minutos.
- H₁: La media de ejercicio semanal ≠ 150 minutos.

Cálculo del valor-p y comparación con α (0.05).

Explicación de errores tipo I y II en el contexto del estudio.


```python
import scipy.stats as st

# Prueba de hipótesis para Horas de sueño
# H0: media = 7
t_stat_sueno, p_val_sueno = st.ttest_1samp(df['Horas_sueno'], 7)

print("Horas de sueño:")
print("t-stat:", round(t_stat_sueno, 2), "p-valor:", round(p_val_sueno, 4))
if p_val_sueno < 0.05:
    print("Se rechaza H0: la media difiere significativamente de 7 horas.\n")
else:
    print("No se rechaza H0: la media no difiere significativamente de 7 horas.\n")

# Prueba de hipótesis para Ejercicio semanal
# H0: media = 150
t_stat_ej, p_val_ej = st.ttest_1samp(df['Ejercicio_min_sem'], 150)

print("Ejercicio semanal:")
print("t-stat:", round(t_stat_ej, 2), "p-valor:", round(p_val_ej, 4))
if p_val_ej < 0.05:
    print("Se rechaza H0: la media difiere significativamente de 150 minutos.\n")
else:
    print("No se rechaza H0: la media no difiere significativamente de 150 minutos.\n")
```

 **Interpretación preliminar**:

- El valor-p indica la probabilidad de observar los datos si la hipótesis nula fuera cierta.
- Si p < 0.05, rechazamos H₀ y concluimos que existe diferencia significativa.
- Si p ≥ 0.05, no tenemos evidencia suficiente para rechazar H₀.

**Errores posibles:**
- Error tipo I: Rechazar H₀ cuando en realidad es verdadera (falso positivo).
- Error tipo II: No rechazar H₀ cuando en realidad es falsa (falso negativo).

En el contexto del estudio:
- Un error tipo I sería concluir que los estudiantes duermen diferente de 7 horas cuando en realidad no es así.
- Un error tipo II sería no detectar una diferencia real en los hábitos de ejercicio respecto a los 150 minutos.

En cuanto al ejercicio semanal, la media ronda los 150 minutos, pero con mayor dispersión que el sueño. Esto indica que algunos estudiantes cumplen con las recomendaciones de actividad física, mientras que otros presentan niveles muy bajos.
