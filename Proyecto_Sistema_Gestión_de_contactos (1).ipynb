## 2. Registro de contactos

En este apartado el objetivo es agregar nuevos contactos, especificando detalles como:

*   Nombre
*   Teléfono
*   Correo
*   Dirección

Para ello, se utilizará una ***clase*** para representar cada contacto como un objeto, aplicando Programación Orientada a Objetos (POO) y *encapsulación* para controlar los datos de la clase mediante métodos. Asimismo, se agrega una clase denominada **gestor de contactos** para generar una lista donde se almacenen los contactos. Considerar que para efectos de estas clases se utilizará el método *init* como constructor de la clase y el método *str* para definir el objeto y otros que se detallarán en cada caso.



   2.1 Definición del Contacto

```python
class Contacto:
  """Clase que representa un contacto individual. Contiene atributos básicos como nombre, teléfono, correo y dirección"""

#se utiliza este  como constructor de la clase asegurando que contacto tenga los atributos respectivos
  def __init__(self, nombre, telefono, correo, direccion):
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo
        self.direccion = direccion

  def __str__(self):
      """Devuelve una representación legible del contacto."""

      return f"{self.nombre} | {self.telefono} | {self.correo} | {self.direccion}"
```

2.2 Agregar nuevo contacto

```python
class Contacto:
  """Clase que representa un contacto individual. Contiene atributos básicos como nombre, teléfono, correo y dirección"""

#se utiliza este  como constructor de la clase asegurando que contacto tenga los atributos respectivos
  def __init__(self, nombre, telefono, correo, direccion):
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo
        self.direccion = direccion

  def __str__(self):
      """Devuelve una representación legible del contacto."""

      return f"{self.nombre} | {self.telefono} | {self.correo} | {self.direccion}"

#Agregar contactos a una lista
class GestorContactos:
  """Clase que gestiona una lista de contactos. En este caso, permite realizar operaciones CRUD (Create, Read, Update, Delete)"""

  def __init__(self):
      # Lista que almacenará objetos de tipo Contacto
        self.contactos = [] # Aquí está la lista

  def agregar_contacto(self, nombre, telefono, correo, direccion):
      """Agrega un nuevo contacto a la lista"""

    #validación del teléfono
      if telefono:
        if not telefono.isdigit():
          print ("El teléfono debe contener sólo números")
          return
        if len(telefono) <8 or len(telefono)>9:
          print("El télefono debe tener entre 8 y 9 dígitos.")
          return
      #si pasa la validación, se crea el contacto

      contacto = Contacto(nombre, telefono, correo, direccion)
      self.contactos.append(contacto) #método para agregar nuevo contacto a la lista de contactos
      print(f"✅ Contacto agregado exitosamente, bienvenid@ {nombre}")

#Ejemplo1
gestor = GestorContactos()
gestor.agregar_contacto("Juan Pérez", "123456789", "juan@mail.com", "Concepción")

#Ejemplo2
gestor = GestorContactos()
gestor.agregar_contacto("José Pérez", "123entel", "jose@mail.com", "Concepción")


```

## 3. Edición y Eliminación de contactos

Por otro lado, parte de los objetivos del proyecto es que los usuarios puedan modificar o eliminar los contactos existentes. Por tanto, como dicho proceso requiere de un bucle para que recorra cada elemento de la lista, se opta por utilizar la iteración **for** asociada a una condición **if**. En otras palabras, para un nombre en específico, se podrá cambiar el teléfono, correo y/o dirección.

Existe la opción de considerar el for con range, no obstante, para ello se debe conocer el rango de la lista a evaluar, es decir, su posición. Sin embargo, solo interesa el objeto como tal, no su posición.

3.1 Edición/Actualización de contacto



```python
  #Edición de contacto; al utilizar "None" que significa vacío, permite al usuario modificar solo lo necesario.
def editar_contacto(self, nombre, nuevo_telefono=None, nuevo_correo=None, nueva_direccion=None):
    """Edita los datos de un contacto existente. Se puede actualizar teléfono, correo y dirección"""
    for  contacto in self.contactos:
      if contacto.nombre.lower() == nombre.lower(): #método que convierte el texto en minúsculas para comparar valores
        if nuevo_telefono:
          contacto.telefono = nuevo_telefono
        if nuevo_correo:
          contacto.correo = nuevo_correo
        if nueva_direccion:
          contacto.direccion = nueva_direccion
        print(f"Contacto actualizado con éxito: {nombre}")
        return

    print("Contacto no encontrado. Revise si está escrito correctamente.")

GestorContactos.editar_contacto = editar_contacto #Se asigna como método de la clase GestorContactos. (útil para  una función fuera de la clase)
```

3.2 Eliminación de contacto

```python
#Elimina contacto usando condicional y bucle for

def eliminar_contacto(self, nombre):
  """Elimina un contacto de la lista por su nombre"""

  for contacto in self.contactos:
          if contacto.nombre.lower() == nombre.lower(): #método que convierte el texto en minúsculas para comparar valores
              self.contactos.remove(contacto) #método que busca un elemento exacto y lo elimina, sino arroja error.
              print(f" Contacto eliminado con éxito: {nombre}")
              return
              print(" Contacto no encontrado, ingresa el nombre correctamente.")

GestorContactos.eliminar_contacto = eliminar_contacto
```

## 4. Búsqueda de contactos

Por último, el proyecto permite buscar contactos por nombre o número de teléfono.

En este caso, es necesario utilizar una función que genere la búsqueda para luego a través de una comprensión de listas genere un resultado, a la cual, se le denomina lista filtrada o de coincidencias.

4.1 Buscar contacto

```python
#Buscar contacto a través de condicional y comprensión de listas

def buscar_contacto(self, criterio_buscado):
  """Busca contactos por nombre o número de teléfono. Devuelve todos los resultados que coincidan con el criterio."""

  resultados = [c for c in self.contactos if criterio_buscado.lower() in c.nombre.lower() or criterio_buscado in c.telefono]
#verifica si existen resultados en la lista

  if resultados: #muestra resultados encontrados
            print("Resultados de búsqueda:")
            for c in resultados:
                print(c)
  else:
            print("No se encontraron contactos. Intente nuevamente.")

GestorContactos.buscar_contacto = buscar_contacto #Se asigna como método de la clase GestorContactos
```

```python
#Mostrar contactos usando bucle for

def mostrar_contactos(self):
      """Muestra todos los contactos registrados en la lista."""

      if self.contactos: #lista que contiene todos los contactos
            print("Lista de contactos:")
            for c in self.contactos:
                print(c)
      else:
            print("No hay contactos registrados.")

GestorContactos.mostrar_contactos = mostrar_contactos #Se asigna como método de la clase GestorContactos
```
